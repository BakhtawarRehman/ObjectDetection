```
# IMport Object Detection module from imageAI 
from imageai.Detection import ObjectDetection
import os

execution_path = '/content/drive/My Drive'
print(execution_path)
```

```
#Loading Detection model.
#Here i have used tinyyyolo
#It is better to use other models.
Detection = ObjectDetection()
Detection.setModelTypeAsTinyYOLOv3()
Detection.setModelPath(os.path.join(execution_path, "yolo-tiny.h5"))
Detection.loadModel()
```

```
detections = []
detections.append(Detection.detectObjectsFromImage(input_image='/content/2.jpg', output_image_path="/content/a.jpg", minimum_percentage_probability=30))
detections.append(Detection.detectObjectsFromImage(input_image='/content/ru.jpg', output_image_path="/content/ruO.jpg", minimum_percentage_probability=10))
```

```
for entity in detections:
  for eachObject in entity:
    print(eachObject["name"] , " : ", eachObject["percentage_probability"], " : ", eachObject["box_points"] )
    print("--------------------------------")
```

Let's try video object detection OR object tracking below

```
from imageai.Detection import VideoObjectDetection
```

```
vDetector = VideoObjectDetection()
vDetector.setModelTypeAsTinyYOLOv3()
vDetector.setModelPath(os.path.join(execution_path, "yolo-tiny.h5"))
vDetector.loadModel()
```

```
video_path = vDetector.detectObjectsFromVideo(input_file_path="/content/drive/My Drive/traffic.mp4",
                                 output_file_path="/content/traffic_detected",
                                 frames_per_second=60, log_progress=True)
```
